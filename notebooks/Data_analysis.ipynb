```python
# Data paths locally
PATH = '/home/marys/Documents/Cell_nuclei_segmentation/data'
RAWIMAGES_PATH = PATH + '/rawimages'
GROUNDTRUTH_PATH = PATH + '/groundtruth'
```

```python
description = pd.read_csv(f'{PATH}/image_description.csv', delimiter=';')
train_data = []
train_masks = []
test_data = []
test_masks = []

for _, r in description.iterrows():
    image_name = r[0]
    train_or_test = r[3]
    if train_or_test == 'train':
        train_data.append(imageio.imread(f'{RAWIMAGES_PATH}/{image_name}.tif'))
        train_masks.append(imageio.imread(f'{GROUNDTRUTH_PATH}/{image_name}.tif'))
    else:
        test_data.append(imageio.imread(f'{RAWIMAGES_PATH}/{image_name}.tif'))
        test_masks.append(imageio.imread(f'{GROUNDTRUTH_PATH}/{image_name}.tif'))

train_data = np.array(train_data)
train_masks = np.array(train_masks)
test_data = np.array(test_data)
test_masks = np.array(test_masks)
```

```python
for i in train_data:
    print(i.shape)
print("---------")
for i in train_masks:
    print(i.shape)
print("---------")
for i in test_data:
    print(i.shape)
print("---------")
for i in test_masks:
    print(i.shape)
```
